```python
import matplotlib.pyplot as plt

# X-axis: Algorithms
algorithms = ['Naive', 'KMP', 'Rabin-Karp', 'Boyer-Moore']

# Y-axis: Number of character comparisons
comparisons = [29, 17, 16, 13]

# Plotting the line graph
plt.figure(figsize=(10, 6))
plt.plot(algorithms, comparisons, marker='o', linestyle='-', color='darkblue', linewidth=2)

# Annotating data points with the comparison values
for i, value in enumerate(comparisons):
    plt.text(i, value + 0.8, str(value), ha='center', fontsize=11, color='black')

# Titles and labels
plt.title('Figure 1: Character Comparisons of String Matching Algorithms', fontsize=14)
plt.xlabel('Algorithms')
plt.ylabel('Total Comparisons')
plt.grid(True, linestyle='--', alpha=0.6)

# Save and display the plot
plt.tight_layout()
plt.savefig("Figure_1_Character_Comparisons_Corrected.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Extended data: pattern sizes vs character comparisons
patterns = [1, 2, 3, 4, 5, 6]
naive = [250, 520, 790, 1040, 1300, 1550]
kmp = [100, 210, 300, 410, 500, 610]
rabin_karp = [120, 230, 350, 470, 590, 700]
boyer_moore = [80, 160, 240, 310, 400, 460]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(patterns, naive, marker='o', linestyle='-', label='Naive', color='red')
plt.plot(patterns, kmp, marker='s', linestyle='--', label='KMP', color='blue')
plt.plot(patterns, rabin_karp, marker='^', linestyle='-.', label='Rabin-Karp', color='green')
plt.plot(patterns, boyer_moore, marker='d', linestyle=':', label='Boyer-Moore', color='purple')

# Annotations
for i in range(len(patterns)):
    plt.text(patterns[i], naive[i] + 30, str(naive[i]), ha='center', fontsize=8, color='red')
    plt.text(patterns[i], kmp[i] + 30, str(kmp[i]), ha='center', fontsize=8, color='blue')
    plt.text(patterns[i], rabin_karp[i] + 30, str(rabin_karp[i]), ha='center', fontsize=8, color='green')
    plt.text(patterns[i], boyer_moore[i] + 30, str(boyer_moore[i]), ha='center', fontsize=8, color='purple')

# Labels and title
plt.title('Figure 2: Character Comparisons vs Number of Patterns', fontsize=14)
plt.xlabel('Number of Patterns')
plt.ylabel('Total Comparisons')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Save & show
plt.tight_layout()
plt.savefig("Figure_1_Character_Comparisons_MultiPattern.png")
plt.show()
```
